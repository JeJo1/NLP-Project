```python
import regex as re
from util import clean
```

```python
train_dir = 'Dataset/train.txt'
val_dir = 'Dataset/val.txt'
test_dir = 'Dataset/test.txt'

train_dump = 'Dataset/train_clean.txt'
val_dump = 'Dataset/val_clean.txt'
test_dump = 'Dataset/test_clean.txt'
```
